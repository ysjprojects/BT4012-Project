```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('../data/3_final_dataset.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
X = df.drop('is_fraud',axis=1)
y = df['is_fraud']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.4, stratify=y)
```

```python
X_train.insert(0,'is_fraud',y_train)
X_test.insert(0,'is_fraud',y_test)
```

```python
X_test.to_csv('../data/4_test_dataset.csv', index=False)
```

```python
X_train.to_csv('../data/5_train_dataset.csv', index=False)
```
