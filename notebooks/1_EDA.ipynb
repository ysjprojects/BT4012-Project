# Importing Packages

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)
```

# Data Reading

```python
df = pd.read_csv('transaction_dataset.csv', index_col=0)
print(df.shape)
df.head()
```

## Getting summary statistics

```python
df.describe().T
```

## Getting column information

```python
df.info()
```

# Exploratory data analysis

## Object columns

```python
object_cols = df.select_dtypes('O').columns
print(f'Columns of object data type: {object_cols.tolist()}')
```

### Unique counts

```python
for object_col in object_cols:
    unique_count = df[object_col].nunique()
    print(f'Column {object_col} has {unique_count} unique values')
```

## Numerical columns

```python
numerical_cols = df.select_dtypes(['int', 'float']).columns
print(f'Columns of numerical data type: {numerical_cols.tolist()}')
```

### Numerical cols with 0 variance

```python
df[numerical_cols].var()
```

```python
numerical_cols_with_0_variance = df[numerical_cols].columns[[df[numerical_cols].var() == 0]].tolist()
print(f'Numerical columns with 0 variance: {numerical_cols_with_0_variance}')
```

### Correlation Matrix

```python
corr_matrix = df[numerical_cols].corr()
corr_matrix.head()
```

```python
masking = np.zeros_like(corr_matrix)
masking[np.triu_indices_from(masking)]=1

plt.figure(figsize=(18, 12))
sns.heatmap(corr_matrix, annot=False, center=0, cmap='Spectral', mask=masking)
plt.title('Heatmap of correlations')
plt.show()
```

## Missing values

```python
num_null_values_by_columns = df.isnull().sum()
print(f'Number of null values for each columns: \n{num_null_values_by_columns}')
```

```python
num_rows = df.shape[0]
percentage_null_values_by_columns = round(num_null_values_by_columns / num_rows * 100, 2)
print(f'Percentage of null values for each columns: \n{percentage_null_values_by_columns}')
```

## Pie chart of target column

```python
flag_value_counts = df['FLAG'].value_counts()
flag_value_counts
```

```python
plt.figure(figsize= (12, 8))
plt.pie(flag_value_counts, autopct= '%.2f%%', labels= ['Not Fruad', 'Fraud'])
plt.title('Percentage of Fraud')
plt.show()
```

## Box Plot

```python
num_numerical_cols = len(numerical_cols)

fig, axes = plt.subplots(num_numerical_cols//3+1, 3, figsize=(12, 12), constrained_layout =True)
plt.subplots_adjust(wspace = 0.7, hspace=0.8)

for k, numerical_col in enumerate(numerical_cols, 0):
    i = k // 3
    j = k % 3
    sns.boxplot(ax = axes[i, j], data=df, x=numerical_col)


plt.show()
```
