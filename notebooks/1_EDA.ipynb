# Exploratory data analysis

## Object columns

```python
object_cols = df.select_dtypes('O').columns
print(f'Columns of object data type: {object_cols.tolist()}')
# df[object_cols]
```

### Unique counts

```python
for object_col in object_cols:
    unique_count = df[object_col].nunique()
    print(f'Column {object_col} has {unique_count} unique values')
```

## Numerical columns

```python
numerical_cols = df.select_dtypes(['int', 'float']).columns
print(f'Columns of numerical data type: {numerical_cols.tolist()}')
```

### Numerical cols with 0 variance

```python
df[numerical_cols].var()
```

```python
numerical_cols_with_0_variance = df[numerical_cols].columns[df[numerical_cols].var() == 0].tolist()
print(f'Numerical columns with 0 variance: {numerical_cols_with_0_variance}')
```

### Numerical cols to be used as features


```python
numerical_feature_list = list(set(df.columns)- set(['Index','Address', 'FLAG']) - set(object_cols) - set(numerical_cols_with_0_variance))
```

### Correlation Matrix

```python
corr_matrix = df[numerical_feature_list + ['FLAG']].corr()

masking = np.zeros_like(corr_matrix)
masking[np.triu_indices_from(masking)]=1

plt.figure(figsize=(18, 12))
sns.heatmap(corr_matrix, annot=False, center=0, cmap='Spectral', mask=masking)
plt.title('Heatmap of correlations')
plt.show()
```

```python
pairs_with_high_corr = []

# Iterate through the upper triangle of the correlation matrix
for i in range(len(corr_matrix.columns)):
    for j in range(i + 1, len(corr_matrix.columns)):
        if corr_matrix.iloc[i, j] > 0.7:
            pairs_with_high_corr.append((corr_matrix .columns[i], corr_matrix .columns[j]))

# Display the pairs with correlation equal to 1
for pair in pairs_with_high_corr:
    print(pair)
    
```

```python
all_variables = [var for pair in pairs_with_high_corr for var in pair]
unique_variables = set(all_variables)


print("Unique Variables from Pairs with Correlation more than 0.7:")
print(unique_variables)
print(len(unique_variables))
```

## Missing values

## Box Plot

```python
numerical_feature_list_num = len(numerical_feature_list)

num_cols_per_row = 3  # Number of numerical columns to display per row
num_rows = (len(numerical_feature_list) + num_cols_per_row - 1) // num_cols_per_row
fig, axes = plt.subplots(num_rows, num_cols_per_row, figsize=(12, 12), constrained_layout =True)
plt.subplots_adjust(wspace = 0.7, hspace=0.8)

for k, col in enumerate(numerical_feature_list, 0):
    i = k // 3
    j = k % 3
    sns.boxplot(ax = axes[i, j], data=df, x=col)

# Remove any unused subplots
for i in range(len(numerical_feature_list), num_rows * num_cols_per_row):
    fig.delaxes(axes.flatten()[i])


plt.show()
```

## Density plot of numerical columns

```python
# Create subplots for each numerical column
num_cols_per_row = 3  # Number of numerical columns to display per row
num_rows = (len(numerical_feature_list) + num_cols_per_row - 1) // num_cols_per_row
fig, axes = plt.subplots(num_rows, num_cols_per_row, figsize=(15, 5 * num_rows))
fig.subplots_adjust(hspace=0.5)

# Plot density graphs for each numerical column
for i, col in enumerate(numerical_feature_list):
    row = i // num_cols_per_row
    col_num = i % num_cols_per_row

    ax = axes[row, col_num]
    sns.kdeplot(df.loc[df['FLAG'] == 0, col], label='Not Fraud', ax=ax, shade = True)
    sns.kdeplot(df.loc[df['FLAG'] == 1, col], label='Fraud', ax=ax, shade = True)
    ax.set_title(col)
    
    # Add a legend
    ax.legend(title='FLAG', labels=['0', '1'])

# Remove any unused subplots
for i in range(len(numerical_feature_list), num_rows * num_cols_per_row):
    fig.delaxes(axes.flatten()[i])

# Display the plots
plt.show()
```
