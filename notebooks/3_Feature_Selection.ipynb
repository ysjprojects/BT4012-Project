```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('../data/2_cleaned_dataset.csv', index_col=0)
df.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix


X = df.drop(['is_fraud'],axis=1)
y = df['is_fraud']

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,stratify=y)

baseline_rf = RandomForestClassifier()
baseline_rf.fit(X_train, y_train)
y_pred = baseline_rf.predict(X_test)
rf_cm = confusion_matrix(y_test, y_pred)
tn, fp, fn, tp = rf_cm.ravel()

accuracy = (tp + tn) / (tn + fp + fn + tp)
recall = tp / (tp + fn)
precision = tp / (tp + fp)
f1 = 2 * precision * recall / (precision + recall)


plt.figure(figsize=(9,9))
sns.heatmap(rf_cm, annot=True, fmt=".3f", linewidths=.5, square = True, cmap = 'Blues_r');
plt.ylabel('Actual label');
plt.xlabel('Predicted label');
all_sample_title = 'Accuracy: {0:.5f}\n Recall: {1:.5f}\n Precision: {2:.5f}\n F1: {3:.5f}'.format(accuracy,recall, precision, f1)
plt.title(all_sample_title, size=15)
plt.show()
```

```python
importances = baseline_rf.feature_importances_
std = np.std([tree.feature_importances_ for tree in baseline_rf.estimators_], axis=0)
forest_importances = pd.Series(importances, index=X_train.columns).sort_values(ascending=False)


fig, ax = plt.subplots(figsize=(25,10))
forest_importances.plot.bar(yerr=std, ax=ax)
ax.set_title("Feature importances using MDI")
ax.set_ylabel("Mean decrease in impurity")
fig.tight_layout()
```

```python
def rank_feature_importances(feature_set):
  rf = RandomForestClassifier()
  rf.fit(X_train[feature_set], y_train)
  importances = rf.feature_importances_
  sorted_importance = pd.Series(importances, index=feature_set).sort_values(ascending=False)

  features_ranked = sorted_importance.index.tolist()
  importance_ranking={}
  for index, feature in enumerate(features_ranked):
    importance_ranking[feature] = index

  return importance_ranking
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
# Variance Inflation Factor (VIF) Inference
# VIF threshold: 2.5; https://corporatefinanceinstitute.com/resources/data-science/variance-inflation-factor-vif/

def get_vif(features, df):
    
    X = df[features]
    # the calculation of variance inflation requires a constant
    X['intercept'] = 1
    
    # create dataframe to store vif values
    vif = pd.DataFrame()
    vif["Variable"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    vif = vif[vif['Variable']!='intercept']
    return vif

def drop_feature(df, importance_ranking):
    vif_df = get_vif(df.columns, df).sort_values('VIF', ascending=False)
    feature0 = vif_df.iloc[0]['Variable']
    feature1 = vif_df.iloc[1]['Variable']
    if importance_ranking[feature0] < importance_ranking[feature1]:
      return vif_df, df.drop(feature1, axis=1)
    return vif_df, df.drop(feature0,axis=1)

def vif_iterate(df, min_features_count, vif_cutoff, verbose=False):
    while len(df.columns) > min_features_count:
        vif_df = get_vif(df.columns,df).sort_values('VIF', ascending=False)
        importance_ranking = rank_feature_importances(df.columns)
        if verbose:
          print(vif_df)
          print()
        if np.sum(vif_df['VIF'] >= vif_cutoff)==0:
            break
        _, df = drop_feature(df, importance_ranking)
    return get_vif(df.columns,df).sort_values('VIF', ascending=False), df

```

```python
vif_df,_= vif_iterate(X_train, 10,2.5)
vif_columns = vif_df['Variable'].reset_index(drop=True).to_numpy()
print(vif_df)
print()
print(vif_columns)
```
