```python
import pandas as pd
from sdv.metadata import SingleTableMetadata
from sdv.single_table import CTGANSynthesizer
from sdv.evaluation.single_table import evaluate_quality
from sdv.sampling import Condition
```

```python
train = pd.read_csv('../data/5_train_dataset.csv')
train.head()
```

```python
metadata = SingleTableMetadata()
metadata.detect_from_dataframe(data=train)
metadata
```

```python
synthesizer = CTGANSynthesizer(metadata,verbose=True)
synthesizer.fit(train)scresc
```

```python
synthesizer.save(
    filepath='../models/ctgan.pkl'
)
```

```python
diff = train['is_fraud'].value_counts()[0] - train['is_fraud'].value_counts()[1]
diff
```

```python
def get_fraud_cond(num_rows):

    return Condition(
        num_rows=num_rows,
        column_values={'is_fraud': 1}
    )

def get_non_fraud_cond(num_rows):

    return Condition(
        num_rows=num_rows,
        column_values={'is_fraud': 0}
    )
```

```python
# Balance the dataset

synthetic_data_fraud = synthesizer.sample_from_conditions(
    conditions=[get_fraud_cond(25000+diff)])

synthetic_data_non_fraud = synthesizer.sample_from_conditions(
    conditions=[get_non_fraud_cond(25000)])
```

```python
synthetic_data = pd.concat([synthetic_data_fraud, synthetic_data_non_fraud])
synthetic_data['is_fraud'].value_counts()
```

```python
quality_report = evaluate_quality(
    real_data=train,
    synthetic_data=synthetic_data_fraud,
    metadata=metadata)
```

```python
quality_report = evaluate_quality(
    real_data=train,
    synthetic_data=synthetic_data_non_fraud,
    metadata=metadata)
```

```python
synthetic_data = synthetic_data.reset_index(drop=True)
```

```python
quality_report = evaluate_quality(
    real_data=train,
    synthetic_data=synthetic_data,
    metadata=metadata)
```
